```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

# Histogramme
Ein Histogramm ist eine flächenproportionale Darstellung der relativen Häufigkeiten.

```python
daten = np.array([5,1,3,3,3,5,1,1,2])
unique, frequency = np.unique(daten, return_counts = True)
print('Merkmal relative Häufigkeit')
for u, f in zip(unique, frequency):
    print(f'{u:<7} {f/len(daten):.3f}')
plt.ioff(); fig = plt.figure(); plt.ion()
n, bins, patches = plt.hist(daten, bins=[1,2,3,5,6], density=True, alpha=0.4, color='g', edgecolor='black')
plt.title('Histogramm')
plt.ylabel('rel. Häufigkeitsdichte')
plt.show()
print('Die Höhe der Balken:', n)
```

```python
plt.ioff(); fig = plt.figure(); plt.ion()
n, bins, patches = plt.hist(daten, bins=[0,2,5,7], density=True, alpha=0.4, color='g', edgecolor='black')
plt.title('Histogramm')
plt.ylabel('rel. Häufigkeitsdichte')
plt.show()
print('Die Höhe der Balken:', n)
```

# Verteilungsfunktion
Die Verteilungsfunktion $F(x)$ gibt für $x\in\mathbb{R}$ den Anteil der Beobachtungswerte an, die kleiner oder gleich $x$ sind.

```python
plt.ioff(); fig = plt.figure(); plt.ion()
plt.hist(daten, bins=unique, density=True, cumulative=True, label='Verteilungsfunktion', histtype='step', alpha=0.8, color='blue')
plt.show()
```
