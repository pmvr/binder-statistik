# Releative und absolute Häufigkeiten

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## Absolute und relative Häufigkeiten einer Datenreihe als Tabelle

```python
daten = np.array([5,1,3,3,3,5,1,1,2])
unique, frequency = np.unique(daten, return_counts = True)
print('Merkmal absolute Häufigkeit')
for u, f in zip(unique, frequency):
    print(f'{u:<7} {f}')
```

```python
print('Merkmal relative Häufigkeit')
for u, f in zip(unique, frequency):
    print(f'{u:<7} {f/len(daten):.3f}')
```

## Absolute und relative Häufigkeiten einer Datenreihe als Balkendiagramm

```python
data = np.array([3, 2, 1])
labels = ['schwarz', 'weiß', 'blau']
plt.xticks(range(len(data)), labels)
plt.ylabel('absolute Häufigkeiten')
plt.title('Farbe der Autos')
plt.bar(range(len(data)), data, alpha=0.4) 
plt.show()
```

```python
fig = plt.figure(figsize=(8,4))
plt.xticks(range(len(data)), labels)
plt.ylabel('relative Häufigkeiten')
plt.title('Farbe der Autos')
plt.bar(range(len(data)), data/sum(data), color='red', alpha=0.4) 
plt.show()
```

## Ergebnis der Bundestagswahl 2013

```python
data = np.array([41.5, 25.7, 8.6, 8.4, 100-41.5-25.7-8.6-8.4])
labels = ['Union', 'SPD', 'Linke', 'Grüne', 'Sonstige']
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
axes[0].set_xticks(range(len(data)), labels)
axes[0].set_ylabel('%')
axes[0].set_title('Bundestagswahl 2013')
axes[0].bar(range(len(data)), data, width=0.5, alpha=0.4) 
axes[1].set_title('Bundestagswahl 2013')
axes[1].pie(data, labels=labels) 
plt.show()
```
