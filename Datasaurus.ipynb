# Das Datasaurus-Dutzend

Hierbei handlet es sich um 13 Datensätze (der Datasaurus + 12 weitere), die alle dieselben statistischen Kennzahlen aufweisen, nämlich
  * x/y-Mittelwlwerte
  * x/y-Varianzen bzw. Standardabweichungen
  * Korrelationskoeffizient von Pearson
  
Die Kennzahlen lauten

| Kennzahl         |  Wert   |
|------------------|---------|
| x-Mittelwert     |  54.26  |
| y-Mittelwert     |  47.83  |
| x-Std-Abweichung |  16.76  |
| y-Std-Abweichung |  26.93  |
| Korrelation      |  -0.06  |

Siehe [Same Stats, Different Graphs: Generating Datasets with Varied Appearance and Identical Statistics through Simulated Annealing](https://www.autodesk.com/research/publications/same-stats-different-graphs).

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

# Daten: https://www.autodesk.com/content/dam/autodesk/www/autodesk-reasearch/Publications/pdf/The%20Datasaurus%20Dozen.zip
df = pd.read_csv('The Datasaurus Dozen/DatasaurusDozen.tsv', sep ='\t')
df
```

```python
fig, axes = plt.subplots(nrows=7, ncols=2, figsize=(15,40))
for i, dataset in enumerate(set(df['dataset'])):
    df[df['dataset'] == dataset].plot.scatter(x='x', y='y', title=dataset, ax=axes[i//2,i%2])
```

```python
# arithmetische Mittel
df.groupby(by="dataset").mean()
```

```python
# Standardabweichung
df.groupby(by="dataset").std()
```

```python
# Korrelation nach Pearson
df.groupby(by="dataset").corr().unstack().iloc[:,1]
```
