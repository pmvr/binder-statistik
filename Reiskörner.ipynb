# Zufällig Verteilung von Reiskörnern



Werden 66 Reiskörner zufällig auf 49 Quadrat verteilt, so ist die resultierende Verteilung näherungsweise Poisson-verteilt.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.special import factorial
from random import randint

anz_quadrate = 49
anz_körner = 66
```

```python
Quadrate = {}
for i in range(1, anz_quadrate+1):
    Quadrate[i] = 0
# Zufallsexperiment
for _ in range(anz_körner):
    Quadrate[randint(1,anz_quadrate)] += 1
df = pd.DataFrame.from_dict(Quadrate, orient='index')
df.plot.bar(figsize=(16,4), title='Anzahl der Reiskörner in den Quadraten');
```

```python
merkmale, häufigkeiten = np.unique(list(Quadrate.values()), return_counts=True)
```

```python
df_hfkt = pd.DataFrame({'Anzahl Körner in einem Quadrat': merkmale, 'Häufigkeiten': häufigkeiten/len(Quadrate)})
df_hfkt.plot.bar(x='Anzahl Körner in einem Quadrat', y='Häufigkeiten');
```

Der Erwartungswert für ein Quadrat beträgt $ \mu = 66/49 \approx 1.35 $

Die Poisson-Verteilung sieht demnach wie folgt aus:

```python
μ = anz_körner/anz_quadrate
k = np.arange(0,6)
df_pois = pd.DataFrame({'Anzahl Körner in einem Quadrat': k, 'Häufigkeiten': μ**k/factorial(k) * np.exp(-μ)})
df_pois.plot.bar(x='Anzahl Körner in einem Quadrat', y='Häufigkeiten');
```
