# Zufllig Verteilung von Reiskrnern



Werden 66 Reiskrner zufllig auf 49 Quadrat verteilt, so ist die resultierende Verteilung nherungsweise Poisson-verteilt.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.special import factorial
from random import randint

anz_quadrate = 49
anz_krner = 66
```

```python
Quadrate = {}
for i in range(1, anz_quadrate+1):
    Quadrate[i] = 0
# Zufallsexperiment
for _ in range(anz_krner):
    Quadrate[randint(1,anz_quadrate)] += 1
df = pd.DataFrame.from_dict(Quadrate, orient='index')
df.plot.bar(figsize=(16,4), title='Anzahl der Reiskrner in den Quadraten');
```

```python
merkmale, hufigkeiten = np.unique(list(Quadrate.values()), return_counts=True)
```

```python
df_hfkt = pd.DataFrame({'Anzahl Krner in einem Quadrat': merkmale, 'Hufigkeiten': hufigkeiten/len(Quadrate)})
df_hfkt.plot.bar(x='Anzahl Krner in einem Quadrat', y='Hufigkeiten');
```

Der Erwartungswert fr ein Quadrat betrgt $ \mu = 66/49 \approx 1.35 $
