# Die Binomialverteilung

Beim idealen Galton-Brett mit $n$ Ebnen fällt eine herabfallende Kugel in jeder Ebene mit der Wahrscheinlichkeit von 50% nach links oder rechts.

![image](files/GaltonBrett.png)

Werden die $n+1$ Auffangbehälter von links nach rechts von $0$ bis $n$ durchnummeriert, und landet eine Kugel im Behälter $k$, dann ist sie $k$-mal nach rechts gefallen. Insgesamt gibt es für eine Kugel $\binom{n}{k}$ Wege, um in den Behälter $k$ zu gelangen.

Die Wahrscheinlichkeit für eine Kugel im Behälter $k$ zu fallen, wird beim idealen Galton-Brett durch die Binomialverteilung mit Erfolgswahrscheinlichkeit $p=0.5$ beschrieben

$$b(n,k) = \binom{n}{k} 0.5^n.$$

## Der Allgemeine Fall

Ist die Erfolgswahrscheinlichkeit bei der Binomialverteilung $p$, so berechnet sich die Warscheinlichkeit für $k$ Erfolge in $n$ Versuchen als

$$b(n,k) = \binom{n}{k} p^k (1-p)^{n-k}.$$

### Frage:
Ein Jäger trifft sein Ziel mit einer Wahrscheinlichkeit von 40%. Mit welcher Wahrscheinlichkeit erzielt er bei 50 Schüssen 22 bis 26 Treffer?

### Lösung

$$ P(22 \le X \le 26) = \sum_{k=22}^{26} b(k,50,0.4) = \sum_{k=22}^{26} \binom{50}{k} \cdot 0.4^k \cdot 0.6^{50-k} $$ 

```python
from scipy.special import binom

# binom(n, k) = n!/(k! * (n-k)!)
P = sum([binom(50,k) * 0.4**k * 0.6**(50-k) for k in range(22,27)])
print("exakte Lösung = %0.1f" % (100*P))
```

```python
from scipy.stats import binom

# binom.pmf(k, n, p, loc=0) Probability mass function, P(X=k)
P = sum([binom.pmf(k, 50, 0.4) for k in range(22,27)])
print("exakte Lösung = %0.1f" % (100*P))
```

```python
# binom.cdf(k, n, p, loc=0) Cumulative distribution function, P(0 <= X <= k)
print("exakte Lösung = %0.1f" % (100*(binom.cdf(26,50,0.4)-binom.cdf(21,50,0.4))))
```
